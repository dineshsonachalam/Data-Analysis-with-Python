```python
# Importing Necessary libraries
import pandas as pd
import datetime
from pandas_datareader import data as web # We are going to pull data from the web.
```

```python
# We create start and end date variables that are date time objects.
start_date = datetime.datetime(2010,1,1)
end_date = datetime.datetime(2015,8,22)
print(type(start_date))
```

```python
print(start_date)
```

```python
print(end_date)
```

```python
# Creating a dataframe
df = web.DataReader('F', 'google',start_date,end_date)
```

```python
print(df.head()) # Printing first 5 rows in the dataframe
```

```python
# pyplot is the basic matplotlib graphing module
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
df['High'].plot() # Choosing only High column from the dataframe
plt.legend()
plt.show()
```

```python
# Choosing Low and High column for plotting
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
df['High'].plot() 
df['Low'].plot() 
plt.legend()
plt.show()
```
